```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
import pandas as pd 
df = pd.read_csv('MFC_Form.csv')
#Extracting some columns
df = df.drop(['Carimbo de data/hora', 'E-mail','Na sua opinião, os temas citados são ou foram abordados na sua formação?','Caso se sinta a vontade, exponha aqui o que você acredita que pode mudar em sua formação.','Comentários:','Obrigada!'], axis=1)
```

```python
df.describe()
```

```python
df_fundamentos = df[df.columns[:8]]
df_abordagem   = df[df.columns[8:14]]
df_gestão      = df[df.columns[14:20]]
df_abord_pess  = df[df.columns[20:32]]
df_abord_comu  = df[df.columns[32:]]
```

```python
figure(figsize=(15,300),dpi=80)
sns.displot(df, kind='kde')
#sns.displot(df['Paradigma sistêmico'], kind='kde')
plt.show()
```

```python
sns.set()
figure(figsize=(15,7),dpi=80)
df_fundamentos.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Médias')
plt.title('Fundamentos ')
plt.show()

figure(figsize=(15,7),dpi=80)
df_abordagem.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Médias')
plt.title('Abordagem Familiar ')
plt.show()

figure(figsize=(15,7),dpi=80)
df_gestão.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Médias')
plt.title('Gestão')
plt.show()

figure(figsize=(15,7),dpi=80)
df_abord_pess.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Médias')
plt.title('Abordagem Centrada na Pessoa')
plt.show()

figure(figsize=(15,7),dpi=80)
df_abord_comu.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Médias')
plt.title('Abordagem Comunitária')
plt.show()

```

```python
df_R1 = df[df['Ano da Residência']=='R1']
df_R2 = df[df['Ano da Residência']=='R2']

#df_remov = df[(df <= 6*df.std())]
```

```python
sns.set()
figure(figsize=(15,7),dpi=80)
#plt.suptitle('Avaliação')
#plt.subplot(2,1,1)
df_R1.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Notas')
plt.title('R1')
plt.show()

figure(figsize=(15,7),dpi=80)
#plt.subplot(2,1,2)
df_R2.mean().plot(kind='bar')
plt.ylim(0,5)
plt.ylabel('Notas')
plt.title('R2')
plt.show()
```
