```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

h = 6.626e-34  # Planck's constant (J s)
c = 3.0e8      # Speed of light (m/s)
k = 1.38e-23   # Boltzmann's constant (J/K)

# Define temperature range
T = np.arange(100, 2001, 100)  # Temperature range from 100 K to 2000 K

# Define wavelength range
lamb = np.arange(1e-9, 7e-6, 1e-9)  # Wavelength range from 1 nm to 3 m

# Define color map based on temperature
colors = cm.jet(np.linspace(0, 1, len(T)))

# Calculate blackbody radiation using Planck's law
for i in range(len(T)):
    B = 2*h*c**2/lamb**5 * 1/(np.exp(h*c/(lamb*k*T[i])) - 1)
    plt.plot(lamb*1e9, B/1e12, color=colors[i], label=str(T[i]) + ' K')

# Add labels and legend
plt.xlabel('Wavelength (nm)')
plt.ylabel('Spectral radiance (W/m$^2$/nm/sr)')
plt.title('Blackbody radiation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.savefig('black_body.pdf')
plt.show()
```
